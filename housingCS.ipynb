```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Load the dataset
data = pd.read_csv('Housing.csv')
```

```python

# Mapping 'yes' and 'no' to 1 and 0 respectively in specified columns
binary_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea']

for col in binary_columns:
    data[col] = data[col].map({'yes': 1, 'no': 0})
```

```python

# 1. Summary Statistics
summary_stats = data.describe()
```

```python

# 2. Data Distribution
plt.figure(figsize=(12, 6))
sns.histplot(data['price'], kde=True)
plt.title('Price Distribution')
plt.show()
```

```python
#3. Grouped Analysis: Average price comparison for properties with and without a basement
grouped_price_by_basement = data.groupby('basement')['price'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(x='basement', y='price', data=grouped_price_by_basement)
plt.title('Average Price with and without Basement')
plt.xlabel('Basement')
plt.ylabel('Average Price')
plt.xticks([0, 1], ['No Basement', 'With Basement'])
plt.show()
```

```python

# 4. Categorical Variables Analysis
categorical_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating',
                       'airconditioning', 'parking', 'prefarea']
for col in categorical_columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data[col])
    plt.title(f'{col.capitalize()} Distribution')
    plt.show()
```

```python

# 5. Relationship Visualization
plt.figure(figsize=(8, 6))
sns.scatterplot(x='area', y='price', data=data)
plt.title('Price vs Area')
plt.show()
```

```python

# 6. Grouped Analysis
grouped_price_by_airconditioning = data.groupby('airconditioning')['price'].median().reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(x='airconditioning', y='price', data=grouped_price_by_airconditioning)
plt.title('Median Price by Air Conditioning')
plt.show()
```

```python

# 7. Missing Values Analysis
missing_values = data.isnull().sum()
```

```python

# 8. Outlier Detection
plt.figure(figsize=(8, 6))
sns.boxplot(x='bedrooms', y='price', data=data)
plt.title('Price Distribution by Bedrooms')
plt.show()
```
